## we have a big problem that in some cities the profit is negative 

```python
negative_profit_cities = df.groupby(["City"]).sum()["Profit"]
negative_profit_cities = negative_profit_cities[negative_profit_cities < 0]
negative_profit_cities
```

```python
negative_profit_cities.sum()
```

```python
positive_profit_cities = df.groupby(["City"]).sum()["Profit"]
positive_profit_cities = positive_profit_cities[positive_profit_cities >= 0]
positive_profit_cities
```

```python
positive_profit_cities.sum()
```

```python
positive_profit_cities.sum()+negative_profit_cities.sum()
```
